### 绘制一个SVD的视觉表示

奇异值分解（SVD）可以将N×K矩阵分解成$U\Sigma V^T$。 在文献中有不同的计算SVD的惯例，这个图表示了本文中使用的惯例。 奇异值$\Sigma$的矩阵总是大小为[R×R]的方阵，其中$R = min（N，K）$。 得到的U和V矩阵的形状取决于N或K是否较大。 矩阵U的列被称为左奇异向量，矩阵V的列被称为右奇异向量。列是正交基，并且满足$U^TU=V^TV=I$。

```python
%run fig_svd_visual.py
```
