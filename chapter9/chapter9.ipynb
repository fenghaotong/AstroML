### 高斯朴素贝叶斯分类

高斯朴素贝叶斯分类方法用于从不可变主序列恒星中分离可变的RR Lyrae恒星。 在左侧面板中，浅灰色的点显示不可变的来源，而黑色的点显示可变的来源。 分类边界由黑线表示，分类概率由阴影背景表示。 在右侧面板中，我们根据拟合中使用的特征数量来显示完整性和污染程度。 对于单个特征，使用u-g。 对于两个特征，u-g和g = r被使用。 对于三个特征，使用u-g，g-r和r-i。 很明显g-r颜色是最好的鉴别器。 所有四种颜色，朴素贝叶斯达到了0.876的完整性和0.790的污染。

```python
%run fig_rrlyrae_naivebayes.py
```
