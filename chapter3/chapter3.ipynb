### 均匀分布值的均值的收敛性

用样本大小范围从N = 100到N = 10,000的比较两个估计量对均匀分布的位置参数的样本大小依赖性。 顶部的估计量是样本均值，底部的估计量是两个极值的平均值。 为了比较，显示了理论上的1，2和3σ曲线。 当使用样本均值来估计位置参数时，不确定性按比例减小到1 / N，并且当使用两个极值的平均值作为1 / N时。 注意两个面板的不同垂直刻度。

两种估计均值$\mu$的方法：

- $\overline{\mu} = mean(x)$ 带有一个尺度为$\frac{1}{\sqrt(N)}$的错误
- $\overline{\mu} = \frac{1}{2}[max(x) + min(x)]$ 带有一个尺度为$\frac{1}{N}$的错误

````python
%run fig_uniform_mean.py
````
