### 欧几里德最小生成树

SDSS"great wall"二维投影的近似欧式最小生成树。 上面板显示输入点，中间面板显示连接它们的树状图。 下面的小图显示了基于这个树形图的聚类，通过去除最大的10％的图边来创建，并且保留其余连接的聚类具有30个或更多个成员。

该图的三个小图显示了斯隆数字巡天（SDSS）的一个星系子集的层次聚类。这个地区被称为“SDSS长城”，并且包含了来自地球约300Mpc（约10亿光年）的数千个星系的延伸集群。顶部面板显示超过8,000个星系在地点（0,0）投影到二维平面的位置。中间的面板显示了在星系位置上的欧氏最小生成树（MST）的树状图表示。通过消除大于给定长度的MST的边缘，我们可以测量在该尺度上的聚类量：这是称为分层聚类的一类模型的一个版本。下图显示了这种聚类方法对3.5Mpc边缘截断的结果，以及适合每个聚类内分布的高斯混合模型。
