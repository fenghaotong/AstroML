### 广义与标准的Lomb-Scargle

信号$y（t）= 10 + sin（2\pi t / P）P = 0.3$的标准和广义Lomb-Scargle周期图的对比，对应于$\omega_{0〜21}$.这个例子对于标准的Lomb-Scargle算法在某种意义上很差的，因为在$y_{true}<10$的时间内没有采样点，导致平均值过高估计。 下图显示了这些数据的Lomb-Scargle和广义Lomb-Scargle周期图; 广义方法将预期峰值恢复为最高峰值，而标准方法则错误地选择了在ω〜17.6的峰值（因为它比$\omega_{0〜21}$处的真实峰值高）。 虚线显示广义周期图中最高峰的1％和5％显着性水平，由1000次bootstrap重采样确定。

上面两个图没有将噪音数据传递给Lomb-Scargle，而是通过了基本的非嘈杂数据。 这导致了对Lomb-Scargle权力的高估。

中间两个图我们看到，对于有噪声的数据，在10个时间段内只能检测到30个点。

在第二个附加图中，我们将基线和点数增加了十倍。 通过这种配置，峰值被检测到。
