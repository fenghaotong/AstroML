### 用卡方分布评估一个模型拟合

用卡方分布评估拟合的好坏，这些数据是对一个恒星光度的一系列观测，已知误差线。我们的模型假定恒星的亮度不变，也就是说，数据中所有的分散都是都是由于测量误差造成的。

$\chi_{dof}^2 \approx 1$：表示模型拟合的比较好的结果   
$\chi_{dof}^2 < 1$：表示误差被高估   
$\chi_{dof}^2 > 1$: 表示误差被低估或模型被错误描述

在最后一个例子中，从数据中可以清楚地看到，这颗恒星的光度随着时间的变化而变化
