### Cross Validation Examples（交叉验证）

最适合的线，显然不适合数据。 换句话说，这种情况下的线性模型具有较高的偏差。

```python
%run fig_cross_val_A.py
```

三种增加复杂性的模型应用于我们的数据集。 d = 2模型与上图中的线性模型一样，具有较高的偏差，并且数据不足。 d = 19模型遭受高方差，并过度适应数据。 d = 3模型是这两个极端之间的一个很好的折衷。

```python
%run fig_cross_val_B.py
```

下图第一个图显示了我们作为多项式d的函数的数据模型的均方根（rms）训练误差和验证误差。 水平虚线表示数据中固有散射的级别。 多项式从3到5的模型使验证均方根误差最小化。 底部显示了训练和交叉验证子样本的贝叶斯信息准则（BIC）。 根据BIC，一个3阶多项式给出了最适合这个数据集。

```python
%run fig_cross_val_C.py
```

其中d = 2和d = 3数据的学习曲线。两个模型对于几个数据点具有高的方差，在训练和验证错误之间的差异中可见。 随着点数的增加，很明显d = 2是一个高偏差模型，不能简单地通过增加训练点来改善。
